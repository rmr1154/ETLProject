```python
import pandas as pd
```

```python
#Files paths
reviews_path="resources/beerreviews.zip"
breweries_path="resources/us-breweries.zip"
    
#Using pandas to read the files as Data Frames
reviews_df=pd.read_csv(reviews_path, encoding="utf-8")
reviews=reviews_df.copy()
reviews.dropna(inplace=True)
breweries_df=pd.read_csv(breweries_path, encoding="utf-8")
breweries=breweries_df.copy()
```

```python
#Displaying the first five rows of the Beer Reviews documents
reviews.head()
```

```python
#Displaying the first five rows of the US Breweries documents
breweries.head()
```

```python
#spliting the address into new columns for the Breweries file and droping the repeated columns
breweries.dropna(inplace=True)
sliced_add = breweries['address'].str.split(",", n = 3, expand = True)
breweries["Address"]=sliced_add[0]
breweries["City"]=sliced_add[1]
breweries["State"]=sliced_add[2]
breweries["Zip code"]=sliced_add[3]
new_breweries=breweries.drop(columns=['address','state'])
new_breweries.head()
```
